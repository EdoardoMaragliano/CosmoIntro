# Computing lab 1: initial cosmology calculations

```python
# import modules:
import matplotlib.pyplot as plt
import numpy as np
import scipy
import scipy.integrate as integrate
```

Uso le densità relative di materia e costante cosmologica e H0. Trascuro neutrini e radiazione.
Devo convertire H0 in 1/Mpc

```python
# define parameters:
Omegam = 0.3
OmegaL = 1 - Omegam
H0 = 70.0 / (299792)
```

```python
# set up the redshift grid where we want to do the calculation:
redshift = np.linspace(0.0, 1000.0, 100)
# since we want to do log plots usually logspace is better:
redshift = np.logspace(-1., 3., 100)
# compute maximum redshift:
redshift_max = np.amax(redshift)
```

```python
# define function computing Hubble rate, cosmic time
def H(z):
    return np.sqrt(H0**2*(Omegam*(1+z)**3+OmegaL))
```

plotto $H(z)$ numericamente e confronto con l'andamento asintotico dato dalla soluzione analitica per z in era dominata da materia

```python
# sanity check plot (with analytic scaling in deep matter domination for comparison)
plt.plot(redshift, H(redshift), label='$H(z)$')
plt.plot(redshift, (redshift/redshift_max)**(3/2)*H(redshift_max), label='$z^{3/2}$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('z')
plt.legend()
```

Voglio calcolare le distanze: $\chi(z)=\int_0^{z^*} \frac{1}{H(z)}dz$

```python
# define function to compute comoving radial distance:
def chi(z):
    return integrate.quad(lambda x: 1./H(x), 0, z)[0]

#quad is the algorithm of the library that integrates the function
#lambda is the definition of a function that only lives in that space, in that scope. 
#Useful if one needs to alias names
#quad returns two arguments: the integral and an estimate of the error
#thus we select only the first ([0]) component of the output
```

```python
# sanity check plot:
plt.plot(redshift, [chi(z) for z in redshift], label='$\chi(z)$')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('z')
plt.legend()
```

If redshift increases, distance increases.
We now re-do the same calculations with the CAMB library

```python
# calculation using CAMB -> look at the CAMB tutorial
import camb
# set parameters to match our cosmological model:
pars = camb.CAMBparams()
pars.set_cosmology(H0=70.0, ombh2=0.05*(70.0/100)**2, omch2=(Omegam-0.05)*(70.0/100)**2);
# set_cosmology wants the PHYSICAL DENSITIES, that is the density parameter multiplied by (H0/100)^2
# camb prende H0 in km/s/Mpc, quindi non serve più dividere per c

# calculate background results:
results = camb.get_background(pars)
# calculate comoving radial distance with CAMB:
chi_true = results.comoving_radial_distance(redshift)
```

```python
# sanity check plot:
plt.plot(redshift, [chi(z) for z in redshift], label='us')
plt.plot(redshift, chi_true, linestyle='--', label='CAMB')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('z')
plt.legend()
```

```python
# the previous plot does not look bad, but it's big numbers, let's check that the code is actually accurate:
chi_us = np.array([chi(z) for z in redshift])
plt.plot(redshift[1:], chi_us[1:]/chi_true[1:] - 1., label='relative accuracy')
plt.xscale('log')
plt.xlabel('z')
plt.legend()
```

This shows that we are below 1% at z \sim 10^3. At earlier times we start entering the radiation dominated epoch. 
In our simplified model we have neglected radiation so we expect to find disagreement between the two.
