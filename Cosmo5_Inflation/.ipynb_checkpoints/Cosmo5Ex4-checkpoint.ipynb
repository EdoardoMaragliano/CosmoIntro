## Solution of coupled Scalar Field and Hubble equations

$H^2=\frac{8\pi G}{3}\rho=\frac{8\pi G}{3}(\frac{1}{2}\dot\phi^2+\frac{1}{2}m^2\phi^2)= \frac{1}{6m^2}(\dot\phi^2+m^2\phi^2)$

$\ddot\phi+3H\dot\phi+V'=0$

$\epsilon=-\frac{\dot{H}}{H^2}$

```python
import numpy as np
from scipy.integrate import solve_ivp
from scipy.integrate import simpson
import matplotlib.pyplot as plt
```

define functions for H and field equation

```python
def H(phi, dphi_dt):
  return np.sqrt( (M*phi)**2 + dphi_dt**2) / (M*np.sqrt(6))

def f(t, y):
  return np.array([ y[1],  - (M**2)*y[0] - 3*H(y[0], y[1])*y[1] ])

##f returns a 2d array: the fist derivative in [0], the second derivative in [1]

def eps(H):
    dH_dt=np.diff(H,1,append=H[H.size-1])
    return -dH_dt/H**2

def eta(phi_dot,phi_dotdot,H):
    return -phi_dotdot/(H*phi_dot)

def V(phi):
    return (phi**2)/6
```

```python
# initial conditions
M=1
phi0 =  25
dphi0_dt = -0.01 

y0 = np.array([phi0, dphi0_dt])
start_time = 0.
stop_time = 100 

##indep variable
time = np.linspace(start_time, stop_time, 400)
# numerical integrator
sol = solve_ivp(f, [start_time, stop_time], y0, method='RK45', t_eval=time)
phi=sol.y[0,:]
dphi_dt=sol.y[1,:]
d2phi_dt2=f(time,np.array([phi,dphi_dt]))[1]
H=H(phi,dphi_dt)
V=V(phi)

epsilon=eps(H)
eta=eta(dphi_dt,d2phi_dt2,H)
```

```python
fig=plt.figure(figsize=(4,10),dpi=100)
fig,axs = plt.subplots(5,1,sharex=False)
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=3,
                    wspace=0.4,
                    hspace=0.4)
axs[0].plot(sol.t,phi)
axs[0].set_title('$\phi$')
axs[0].set_xlabel('t')
axs[1].plot(sol.t, H)
axs[1].set_title('$H$')
axs[1].set_xlabel('t')
axs[2].plot(sol.t, epsilon)
axs[2].set_title('$\epsilon$')
axs[2].set_xlabel('t')
axs[3].plot(sol.t, eta)
axs[3].set_title('$\eta$')
axs[3].set_xlabel('t')
axs[4].plot(sol.t, V)
axs[4].set_title('$V$')
axs[4].set_xlabel('t')

title='phi_'+str(phi0)+'-der_'+str(dphi0_dt)+'.pdf'
plt.savefig(title,bbox_inches="tight")
```
